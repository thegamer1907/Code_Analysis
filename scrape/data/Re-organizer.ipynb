```python
import os
import shutil
```

```python
try:
    os.mkdir('Organized')
except:
    pass
excep = ["./.ipynb_checkpoints","./","./Organized"]
```

```python
repeat = 1
i = 1
for root, dirs, files in os.walk('./'):
    if root not in excep:
        for file in files:
            if ".json" in file:
                username = file[:-5]
                new_dir = ""
                try:
                    new_dir = os.path.join('Organized/',username)
                    os.mkdir(new_dir)
                    if username[-1]=='.':
                        new_dir = os.path.join('Organized/',username.strip('.'))
                except:
                    new_dir = os.path.join('Organized/','Repeat'+str(repeat))
                    os.mkdir(new_dir)
                    repeat += 1
                print('Copied',i,'sets',username)
                shutil.copy(os.path.join(root,file),new_dir)
                shutil.copy(os.path.join(root,username+"_A.cpp"),new_dir)
                shutil.copy(os.path.join(root,username+"_TLE.cpp"),new_dir)
                i+=1
print("No. of repeats:",repeat-1)
```
