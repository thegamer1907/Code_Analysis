```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('compiled.csv',header=None)
le = pd.read_csv('Legend.csv',header=None, error_bad_lines=False)
```

```python
arr = df[1].values
```

```python
d = {}
for each in arr:
    l = each.split(',')
    for each2 in l:
        if each2 in d.keys():
            d[each2]+=1
        else:
            d[each2] = 1
```

```python
d
```

```python
plt.bar(range(len(d)), list(d.values()), align='center')
plt.xticks(range(len(d)), list(d.keys()))
plt.show()
```

```python
le.loc[4,1]
```

```python
D = {}
for key,val in d.items():
    D[le.loc[int(key),1]] = val
D
```

```python
fig = plt.figure(figsize=(20, 8))
plt.bar(range(len(D)), list(D.values()), align='center')
plt.xticks(range(len(D)), list(D.keys()),rotation=70)
fig.show()
```

```python
fig = plt.figure(figsize=(20, 18))
plt.pie(list(D.values()),labels=list(D.keys()))
fig.show()
```
