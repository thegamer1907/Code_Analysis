```python
import requests
from bs4 import BeautifulSoup
import os
import json
import time
import calendar
import traceback
import pickle
errcnt = 0
globalcnt = 0
pickle_list = []
```

```python
def html(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content,'html.parser')
    return soup


def processproblem(link,problink,exe,mem,usertype,username):
    global errcnt
    global globalcnt
    page = html(link)
    pre = page.find_all('pre',class_='prettyprint lang-cpp program-source')
    code =  pre[0].text
    te = code.strip()
    ext = '.cpp'
    try:
        fname = str(calendar.timegm(time.gmtime()))
        filename = 'E:\Codes\Code Analysis\contest' + '\\' + fname + ext
        filename2 = 'E:\Codes\Code Analysis\contest' + '\\' + fname + '.json'
        f = open(filename,"w")
        f.write(code)
        f.close()
        globalcnt+=1
        tdict = {}
        tdict["problemlink"] = problink
        tdict["sollink"] = link
        tdict["exec_time"] = exe
        tdict["memory"] = mem
        tdict["usertype"] = usertype
        tdict["username"] = username
        g = open(filename2,"w")
        json.dump(tdict,g)
        g.close()
    except Exception as e:
        print(str(e))
        errcnt+=1




def eachpage(status_page):
    global globalcnt
    global pickle_list
    status_table = status_page.findAll('table',class_='status-frame-datatable')
    tr = status_table[0].findAll('tr')
    tr.pop(0)
    for i in tr:
        td = i.findAll('td')
        sollink = 'https://www.codeforces.com' + td[0].a['href']
        if sollink in pickle_list:
            globalcnt+=1
            continue
        usertype = td[2].a['title'].split(' ')[0]
        username = td[2].a.text.strip()
        problemname = td[3].a.text.split('-')[-1].strip()
        lang = td[4].text.strip()
        exe = td[6].text.strip()
        mem = td[7].text.strip()
        problink = 'https://www.codeforces.com' + td[3].a['href']
        status = td[5].span['submissionverdict'].strip()
#         if usertype != 'Unrated,' and usertype != 'Pupil' and usertype != 'Specialist' and usertype != 'Newbie':
#             print(status,usertype,problemname,name,lang)
        if lang == 'GNU C++17' or lang == 'GNU C++14' or lang == 'GNU C++11' or lang == 'GNU C++':
            if status == 'OK':
                pickle_list.append(sollink)
                processproblem(sollink,problink,exe,mem,usertype,username)
                print('done with',globalcnt)





def cfpage():
    global pickle_list
    with open('pickle_list.pkl', 'rb') as f:
        pickle_list = pickle.load(f)
    contest_url = 'https://codeforces.com/contest/868/status'
    status_page = html(contest_url)
    pagination = status_page.findAll('div',class_='pagination')
    li = pagination[1].ul.findAll('li')[-2]
    totpage = int(li.span.a.text)
    for i in range(317,totpage+1):
        pageurl = contest_url + '/page/' + str(i) + '?order=BY_JUDGED_DESC'
        status_page = html(pageurl)
        eachpage(status_page)
        print('page done',i)
```

```python
try:
    cfpage()
except Exception as e:
    print(str(e))
    traceback.print_tb(e.__traceback__)
    with open('pickle_list.pkl', 'wb') as f:
        pickle.dump(pickle_list, f)
```
