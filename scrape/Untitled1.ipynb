```python
from bs4 import BeautifulSoup
import requests
import os
import json
import time
import calendar
import traceback
import pickle
```

```python
a_dict = dict()
tle_dict = dict()

final_dict = dict()
errcnt = 0
```

```python
def insert_a(key,value):
    if key not in a_dict:
        a_dict[key] = value
        if key in tle_dict:
            final_dict[key] = [value,tle_dict[key]]
            
def insert_tle(key,value):
    if key not in tle_dict:
        tle_dict[key] = value
        if key in a_dict:
            final_dict[key] = [a_dict[key],value]



def eachpage(status_page):
    status_table = status_page.findAll('table',class_='status-frame-datatable')
    tr = status_table[0].findAll('tr')
    tr.pop(0)
    for i in tr:
        td = i.findAll('td')
        sollink = 'https://www.codeforces.com' + td[0].a['href']
        usertype = td[2].a['title'].split(' ')[0]
        username = td[2].a.text.strip()
        problemname = td[3].a.text.split('-')[-1].strip()
        lang = td[4].text.strip()
        exe = td[6].text.strip()
        mem = td[7].text.strip()
        problink = 'https://www.codeforces.com' + td[3].a['href']
        status = td[5].span['submissionverdict'].strip()
        if lang == 'GNU C++17' or lang == 'GNU C++14' or lang == 'GNU C++11' or lang == 'GNU C++':
            if status == 'OK':
                insert_a(username  + '+' + problemname,[problink,sollink])
            elif status == 'TIME_LIMIT_EXCEEDED':
                insert_tle(username  + '+' + problemname,[problink,sollink])





def cfpage():
    contest_url = 'https://codeforces.com/contest/1191/status'
    status_page = html(contest_url)
    pagination = status_page.findAll('div',class_='pagination')
    li = pagination[1].ul.findAll('li')[-2]
    totpage = int(li.span.a.text)
    print(totpage)
    for i in range(1,totpage+1):
        pageurl = contest_url + '/page/' + str(i) + '?order=BY_JUDGED_DESC'
        status_page = html(pageurl)
        eachpage(status_page)
        print('page done',i)
        print(len(final_dict))
```

```python
def html(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content,'html.parser')
    return soup
```

```python
def returnCode(link):
    link = 'http' + link.split('ttps')[1]
    page = html(link)
    pre = page.find_all('pre')
    code =  pre[0].text
    te = code.strip()
    table = page.find_all('table')[0].find_all('tr')[1].find_all('td')[7].text
    time = table
    return [te,time]

def writeFile(filename,content):
    f = open(filename,"w")
    f.write(content)
    f.close()

def processproblem(A_link,TLE_link,problink,username,problemName):
    A_code = returnCode(A_link)
    TLE_code = returnCode(TLE_link)
    ext = '.cpp'
    fname = str(calendar.timegm(time.gmtime()))
    A_filename = '/home/harshit/Desktop/Git/Code_Analysis/scrape/data' + '/' + problemName + '/' + username + '_A' + ext
    TLE_filename = '/home/harshit/Desktop/Git/Code_Analysis/scrape/data' + '/' + problemName + '/' + username + '_TLE' + ext
    filename2 = '/home/harshit/Desktop/Git/Code_Analysis/scrape/data' + '/' + problemName + '/'+ username + '.json'
    writeFile(A_filename,A_code[0])
    writeFile(TLE_filename,TLE_code[0])
    tdict = {}
    tdict["problemlink"] = problink
    tdict["A_link"] = A_link
    tdict["TLE_link"] = TLE_link
    tdict["username"] = username
    tdict["A_time"] = A_code[1]
    tdict["TLE_time"] = TLE_code[1]
    g = open(filename2,"w")
    json.dump(tdict,g)
    g.close()
    
```

```python
cnt = 0
for i in final_dict:
    try:
        d = i
        tmp = final_dict[d]
        processproblem(tmp[0][1],tmp[1][1],tmp[0][0],d.split('+')[0],d.split('+')[1])
        cnt+=1
        print(cnt)
    except Exception as e:
        print(str(e))
        errcnt+=1
```
