```python
import requests
from bs4 import BeautifulSoup
import os

globalcnt = 3092

def html(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content,'html.parser')
    return soup

def processproblem(link):
    global globalcnt
    page = html(link)
    pre = page.find_all('pre',class_='prettyprint lang-cpp program-source')
    code =  pre[0].text
    te = code.strip()
    ext = '.cpp'
    try:
        filename = 'E:\Codes\Code Analysis\Codes' + '\\' + str(globalcnt) + ext
        globalcnt+=1
        f = open(filename,"w")
        f.write(code)
        f.close()
    except:
        print('error')


def eachpage(status_page,name):
    status_table = status_page.findAll('table',class_='status-frame-datatable')
    tr = status_table[0].findAll('tr')
    tr.pop(0)
    for i in tr:
        td = i.findAll('td')
        sollink = 'https://www.codeforces.com' + td[0].a['href']
        usertype = td[2].a['title'].split(' ')[0]
        problemname = td[3].a.text.split('-')[-1].strip()
        lang = td[4].text.strip()
        status = td[5].span['submissionverdict'].strip()
        if usertype != 'Unrated,' and usertype != 'Pupil' and usertype != 'Specialist' and usertype != 'Newbie':
#             print(status,usertype,problemname,name,lang)
            if lang == 'GNU C++17' or lang == 'GNU C++14' or lang == 'GNU C++11' or lang == 'GNU C++':
                if status == 'OK' and problemname == name:
                    processproblem(sollink)
                    print(name,problemname)
    
    
                
                
    
def cfpage(page,name):
    contest_table = page.findAll('table',class_='rtable ')
    contest_url = contest_table[0].tbody.tr.a['href']
    contest_url = 'https://www.codeforces.com' + contest_url + '/status'
    status_page = html(contest_url)
    pagination = status_page.findAll('div',class_='pagination')
    li = pagination[1].ul.findAll('li')[-2]
    totpage = int(li.span.a.text)
    for i in range(1,min(totpage,200)):
        pageurl = contest_url + '/page/' + str(i) + '?order=BY_JUDGED_DESC'
        status_page = html(pageurl)
        eachpage(status_page,name)
        print('page done',i)
        
    
```

```python
stopurl = 'https://www.stopstalk.com/problems/tag?page=1&generalized_tags=Binary+Search&site=CodeForces&orderby=solved-count-desc'

stopsoup = html(stopurl)
main_table = stopsoup.find_all('table',class_='bordered centered')
tbody = main_table[0].tbody

tr = tbody.findAll('tr')

for j in range(20,len(tr)):
    i = tr[j]
    td = i.findAll('td')
    name = td[0].a.text.strip()
    purl = td[1].a['href']
    ppage = html(purl)
    cfpage(ppage,name)
    print('DONE WITH A PROBLEM')



```
