```python
import subprocess
import time
import calendar
import os
import shutil
from glob import glob
import json
import datetime
```

```python
folderList = glob("./data/Organized/*/")
```

```python
folderList
```

```python
total = len(folderList)
counter = 0

for folder in folderList:
    jsonFile = glob(folder + '*.json')[0]
    acceptedFile = glob(folder + '*_A.cpp')[0]
    tleFile = glob(folder + '*_TLE.cpp')[0]
    folderName = os.path.basename(os.path.dirname(folder))
    
    with open(jsonFile, 'r') as f:
        j = json.load(f)
        
    starttime = j['TLE_time']
    endtime = j['A_time']
    endtimeObj = datetime.datetime.strptime(endtime, "%Y-%m-%d %H:%M:%S")
    starttimeObj = datetime.datetime.strptime(starttime, "%Y-%m-%d %H:%M:%S")
    
    timeDelta = str(endtimeObj - starttimeObj)
    
    subprocess.call('python ./diff.py {} {} --no-browse --timediff {} -f {}'.format(tleFile, acceptedFile, timeDelta, './data/diffReports/' + folderName + '_report.html'),cwd = os.getcwd())
    
    counter += 1
    
    print("Done %s - %d / %d" % (folderName, counter, total))
```
